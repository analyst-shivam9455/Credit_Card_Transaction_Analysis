# Credit_Card_Tranction_Analysis

```python
# importing necessary libraries
import pandas as pd
import datetime
```

```python
# loading dataset
df = pd.read_csv('credit_card.csv')
df2 = pd.read_csv('Customer.csv')
```

```python
df.head()
```

```python
# converting Week_Start_Date into datetime
df['Week_Start_Date'] = pd.to_datetime(df['Week_Start_Date'],format='mixed')
```

```python
# extracting week from date
df['Week_Number'] = df['Week_Start_Date'].dt.weekday
```

```python
df['Revenue'] = df['Annual_Fees'] + df['Total_Trans_Amt'] + df['Interest_Earned']
```

```python
# save clean data
df.to_csv('Credit_Card_cleaned_data.csv')
```
